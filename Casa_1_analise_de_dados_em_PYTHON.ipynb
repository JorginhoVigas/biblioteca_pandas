```python
import pandas as pd
```

```python
aluguel_casa = pd.read_excel("/content/houses_rent.xlsx")
```

```python
aluguel_casa.head(10) #os 10 primeiros da lista
```

```python
aluguel_casa.tail(10) #os 10 utlimos da lista
```

```python
aluguel_casa = aluguel_casa.rename(columns={"parking spaces":"parking_spaces","rent amount":"rent_amount","property tax":"property_tax","fire insurance":"fire_insurance"}) #editando nomes das colunas
```

```python
aluguel_casa.head()
```

```python
aluguel_casa.dtypes #verificando o tipo de dados das colunas
```

```python
aluguel_casa = aluguel_casa.astype({"rent_amount":"float","property_tax":"float","fire_insurance":"float","total":"float"}) #alterando o tipo de dados das colunas
aluguel_casa.dtypes
```

```python
aluguel_casa.info()
```

```python
aluguel_casa[["area", "rooms", "bathroom", "parking_spaces", "rent_amount", "property_tax", "fire_insurance", "total"]].describe() #verificando questoes de soma
```

```python
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000)] #aplicando filtros
aluguel_casa.head() #imprimindo o valor
```

```python
aluguel_casa.describe()
```

```python
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000) & (aluguel_casa["animal"]=="accept")] #aplicando filtros
aluguel_casa.head() #imprimindo o valor
```

```python
aluguel_casa = aluguel_casa.query("total <= 4000 and animal == 'accept'") #aplicando filtros com query
aluguel_casa.head() #imprimindo o valor
```

```python
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000) & (aluguel_casa["animal"]=="accept") & ((aluguel_casa["rooms"]==2)|(aluguel_casa["rooms"]==3))] #aplicando filtros
aluguel_casa.head() #imprimindo o valor
```

```python
aluguel_casa = aluguel_casa.query(("total <= 4000 and animal == 'accept' and rooms == 2 or rooms == 3")) #aplicando filtros com query
aluguel_casa.head() #imprimindo o valor
```

```python
aluguel_casa.describe()
```

```python
agg_city_price = aluguel_casa.groupby(["city"])["total"].agg("mean").reset_index() #aplicando agrupamento
agg_city_price #imprimindo o valor
```

```python
agg_bath_room_price = aluguel_casa.groupby(["bathroom","rooms"])["total"].agg("mean").reset_index() #aplicando agrupamento
agg_bath_room_price #imprimindo o valor
```

```python
import seaborn as sns #importando biblioteca de graficos
```

```python
ax = sns.barplot(data=agg_bath_room_price, x="rooms", y="total", color = "blue") #imprimindo graficos
ax.set(xlabel = "qtd quartos", ylabel = "media do preco"); #editando os nomes
```

```python
ax = sns.barplot(data=agg_bath_room_price, x="bathroom", y="total", color = "green") #imprimindo graficos
ax.set(xlabel = "qtd banheiros", ylabel = "media do preco"); #editando os nomes
```
