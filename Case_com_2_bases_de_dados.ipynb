```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel("varejo.xlsx") #carregando o arquivo
vendas.head() #imprimindo na tela
```

```python
vendas.info()
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique() #verificando dentro de uma coluna
```

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP","Aplicativo" ) #mudando o nome da coluna
vendas.groupby("idcanalvenda").idcompra.nunique()  #verificando dentro de uma coluna
```

```python
vendas.groupby("estado").idcompra.nunique()  #verificando dentro de uma coluna
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ","_") #mudando o nome da coluna
vendas.groupby("Nome_Departamento").idcompra.nunique()  #verificando dentro de uma coluna
```

```python
vendas[vendas["estado"].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna() # Vendas sem nulo
vendas_sem_nulos.info()
```

```python
vendas["estado"].fillna("MS", inplace=True) #Substituindo o nulo por MS
```

```python
vendas.info() #Verificando se existe algum valor nulo
```

```python
media_preco = vendas["Preço"].mean()  #Pegando a media dos precos

vendas["Preço"].fillna(media_preco, inplace=True)   # Trocando o valor nulo dos precos para a media
```

```python
media_preco
```

```python
vendas.info()
```

```python
mediana_preco = vendas["Preço"].median()
mediana_preco
```

```python
vendas.head()
```

```python
vendas[["Preço", "Preço_com_frete"]].describe() #puxando determinadas colunas e os seus dados
```

```python
preco_errado = vendas.query("Preço > Preço_com_frete") #Filtrando

vendas_correto = vendas.query("Preço < Preço_com_frete") #Filtrando
```

```python
vendas_correto.describe()
```

```python
vendas_correto.head()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe() #Filtrando
```

```python
vendas_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()    #Contando de departamentos mais vendidos
```

```python
round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2) #Media de preco para cada departamento
```

```python
vendas_correto.groupby("Data").idcompra.nunique().sort_values(ascending=False) #Filtrando pela data
```

```python
vendas_correto["mes"] = vendas_correto["Data"].dt.month #gerando a variavel mes
vendas_correto.head()
```

```python
vendas_correto.groupby("mes").idcompra.nunique().sort_values(ascending=False) #Filtrando pelo mes
```

```python
cliente = pd.read_excel("cliente_varejo.xlsx") #importando outro banco de dados
cliente.head()
```

```python
vendas_correto.head()
```

```python
cliente[["idade", "renda"]].describe()
```

```python
cliente.info() #verificando dados das tabelas
```

```python
cliente = cliente.astype({"renda":"float"})  #transformando em float
```

```python
cliente.info() #verificando dados das tabelas
```

```python
vendas_cliente = vendas_correto.merge(cliente, how= "left", on="cliente_Log") #left join no python
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idade_bandeira["bandeira"], agg_idade_bandeira["idade"], color="green")

plt.ylabel("Média de idade")

plt.title("Idade média por Bandeira", loc="center", fontsize=18, fontweight="bold", color= "black");
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"], color="green")

plt.ylabel("Renda média")

plt.title("Renda média por canal de venda", loc="center", fontsize=18, fontweight="bold", color="black");
```

```python
venda_por_data = vendas_correto.groupby("Data").idcompra.nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize=(15,4))

plt.xlabel("Data da venda")
plt.ylabel("Quantidade de vendas")
x = venda_por_data["Data"]
y = venda_por_data["idcompra"]

plt.plot(x,y, color = "green", linewidth = "2")
plt.show()
```

```python
agg_dept_preco = round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2) #Media de preco para cada departamento
agg_dept_preco
```

```python
fig = plt.figure(figsize=(50,25))

plt.bar(agg_dept_preco["Nome_Departamento"], agg_dept_preco["Preço_com_frete"], color="green")

plt.ylabel("Preço média")

plt.title("Preço médio por departamento", loc="center", fontsize=18, fontweight="bold", color="black");
```
